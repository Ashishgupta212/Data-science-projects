# DEALING WITH  MISSING VALUE

```python
covid19_data = Covid_data.replace(to_replace='None', value=np.nan) #  using replace function to convert the 'none' value into nan value in dataset.
```

```python
covid19_data['Corona'].replace(to_replace='other', value=np.nan, inplace=True) #using replace function  to convert the 'other' value into nan value in dataset.
covid19_data.dropna(subset=['Corona'], inplace=True)
```

```python
''' All columns contain four unique values, but these values are identical 
except for the capitalization (one with capital letters and the other with lowercase letters)'''
# To convert all values in a lower letter.

covid19_data['Cough_symptoms'] =  covid19_data['Cough_symptoms'].str.lower()
covid19_data['Fever'] =  covid19_data['Fever'].str.lower()
covid19_data['Sore_throat'] =  covid19_data['Sore_throat'].str.lower()
covid19_data['Shortness_of_breath'] =  covid19_data['Shortness_of_breath'].str.lower()
covid19_data['Headache'] =  covid19_data['Headache'].str.lower()
```

```python
covid19_data.isnull().sum() # how here you can see all null values  in a datasets. 
```

```python
missing_per = (covid19_data.isnull().sum()/len(covid19_data)) *100 # checking the percentage of null values .
missing_per
```

```python
covid19_data=covid19_data.dropna(subset = ['Sore_throat','Shortness_of_breath','Headache'],axis = 0) # delete null value because there very low percentage.
```

```python
covid_data.drop(columns=['Ind_ID','Test_date'],axis =  1,inplace = True)
```

```python
x = covid_data.iloc[:,1:18]
y = covid_data.iloc[:,0]
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

chi_best  = SelectKBest(score_func = chi2 ,k =8)  
k_best =  chi_best.fit(x,y) # CHI SQUARE TEST

np.set_printoptions(precision=3) # PRECISION VALUE MEANS NUMBER OF  DECIMAL .
print(k_best.scores_)


k_feature = k_best.transform(x)
print(k_feature)

selected_columns = x.columns[k_best.get_support()]
print(selected_columns)
```

```python
x =  x.drop(columns=['Cough_symptoms_false', 'Fever_false','Sore_throat_false','Shortness_of_breath_false',
       'Headache_false', 'Age_60_above_No',
       'Age_60_above_Yes','Sex_male', 'Sex_female'],axis =1)
```

```python
x.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2 ,random_state = 51)
```

```python
x.columns
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
scaler =  StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
scaler_data = pd.DataFrame(x_train,columns =['Cough_symptoms_true', 'Fever_true', 'Sore_throat_true',
       'Shortness_of_breath_true', 'Headache_true', 'Known_contact_Abroad',
       'Known_contact_Contact with confirmed', 'Known_contact_Other'])
scaler_data.head()
```

```python
lg =  LogisticRegression()
lg.fit(x_train,y_train)
```

```python
pred =  lg.predict(x_test)
print(pred)
```

```python
score = lg.score(x_test,y_test)
print(score)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(lg,x_train,y_train,cv =5)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
cfr = classification_report(y_test,pred)
print('CLASSIFICATION REPORT:-\n\n',cfr)
```
